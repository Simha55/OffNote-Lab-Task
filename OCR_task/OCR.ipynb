```python
video = cv2.VideoCapture('lyrics.mp4')
```

```python
idx = 0
with open('lyrics.txt','w') as file:
    while video.isOpened():
        ret,frame = video.read()
        if not ret:
            break
     
        if(idx%300 == 0): 
            text = tess.image_to_string(frame, lang = 'eng')
            print(text)
            
            file.write(text + str('\n'))
  
        idx = idx +1
    video.release()
    cv2.destroyAllWindows()
```
